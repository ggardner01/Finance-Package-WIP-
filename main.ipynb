```python
def readRobinHood(sym):
    df = pd.DataFrame(index=web.DataReader("SPY", "robinhood").index.levels[1])
    for s in sym:
        try:
            frp = web.DataReader(s, "robinhood")
            ret = (pd.to_numeric(frp.close_price.values) - pd.to_numeric(frp.open_price.values))/pd.to_numeric(frp.open_price.values)
            df[s] = ret
        except:pass
    return df
```

```python
def ARIMAreturns(df, p=1, d=0, q=0):
    new = pd.DataFrame()
    for v in list(df.columns.values):
        regr = ARIMA(endog = df[v].values, order = (p,d,q))
        pred = regr.fit(method="css").predict(end = 364)
        r = e**(np.sum(pred)) - 1
        new[v] = np.array([r])
    return new
```
